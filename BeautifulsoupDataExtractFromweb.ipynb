```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

r = requests.get('https://jsonplaceholder.typicode.com/posts').json()
df = pd.DataFrame(r)
df.head()


```

```python
response = requests.get('https://www.iana.org/domains/reserved')
html = response.content.decode()
html
```

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
tabledata = soup.find('table')
tablerow = tabledata.find_all('tr')

datarow = []

for i in tablerow:
    cellelement = i.find_all(lambda tag:tag.name in ['th','td'])
    cellvalues = [str(e.string) for e in cellelement]
    datarow.append(cellvalues)

tabledf = pd.DataFrame()
tabledf
```

```python
tabledf.columns = tabledf.iloc[0] #convert first row as column 
tabledf.drop(index = df.index[0],axis=0,inplace=True) # drop thye first index in df
tabledf
```
